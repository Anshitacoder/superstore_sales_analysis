```python
data["Sales"]=data["Sales"].fillna(data["Sales"].median())
```

```python
data.isnull().sum()
```

```python
data["Customer ID"]=data["Customer ID"].fillna("Unknown")
```

```python
data["Region"]=data["Region"].fillna("Unknown")
```

```python
data["Postal Code"]=data["Postal Code"].fillna(0)
```

```python
data.isnull().sum()
```

```python
data=data.drop_duplicates()
```

```python
data.info()
```

```python
print(data.columns.tolist())
```

```python
data.columns=data.columns.str.strip()
```

```python
print(data.columns.tolist())
```

```python
data["Order_Month"]=data["Order"].dt.month
```

```python
data["Order_Year"]=data["Order"].dt.year
```

```python
import seaborn as sb
import matplotlib.pyplot as plt
sb.countplot(x="Category",data=data)
plt.title("Category Distribution")
plt.show()
```

```python
### Insights: office supplies has highest number of order
### Technology has the lowest number of orders
### this indicates that office supplies products are purchased more frequently
```

```python
data.groupby("Region")["Sales"].sum().plot(kind="bar")
plt.title("Sales by Region")
plt.show()
```

```python
### west region has highest sales among all regions
### south region has the lowest sales
### This indicates thet the west region is the strongest market in terms of revenue
```

```python
top_customer=data.groupby("Customer Name")["Sales"].sum().sort_values(ascending=False).head(5)
top_customer.plot(kind="bar")
plt.title("Top 5 Customers")
plt.show()
```

```python
data.groupby("Order_Month")["Sales"].sum().plot()
plt.title("Monthly Sales Trend")
plt.show()
```

```python
### Insights:
###- Sales demonstrate an upward trend over the months, indicating business growth
###- Early months have comparatively lower sales performance
###- The highest sales are observed in the later months, suggesting seasonal demand or year-end purchasing behavior
```
